```python
import instaloader
import pandas as pd
import datetime

# Initialize Instaloader
loader = instaloader.Instaloader()

# Specify the username of the profile you want to analyze
profile_name = 'onefootball'  # Replace with the target Instagram username

# Load the profile
profile = instaloader.Profile.from_username(loader.context, profile_name)

# List to store the data
posts_data = []

# Loop through the last 10 posts
for post in profile.get_posts():
    # Collect data
    post_info = {
        'Post': post.shortcode,
        'Likes': post.likes,
        'Comments': post.comments,
        'Date': post.date_utc,
    }
    posts_data.append(post_info)

    # Stop after 10 posts
    if len(posts_data) >= 10:
        break

# Create a DataFrame from the collected data
df = pd.DataFrame(posts_data)

# Save the DataFrame to a CSV file
df.to_csv(f'{profile_name}_last_10_posts.csv', index=False)

print(f'Data for the last 10 posts of {profile_name} has been saved to {profile_name}_last_10_posts.csv')
df
```
